```python
import earthkit.data
import earthkit.plots
import earthkit.regrid
from polytope.api import Client
```

```python
# Defaults to making a live data request. Set to false to use the cached GRIB file instead.
import os

LIVE_REQUEST = os.getenv("LIVE_REQUEST", "true").lower() == "true"
LIVE_REQUEST
```

```python
request = {
    'activity': 'scenariomip',
    'class': 'd1',
    'dataset': 'climate-dt',
    'date': '20300101',
    'time': "0000",
    'experiment': 'SSP3-7.0',
    'expver': '0001',
    'generation': '1',
    'levtype': 'sfc',
    'model': 'IFS-NEMO',
    'param': '167',
    'realization': '1',
    'resolution': 'high',
    'stream': 'clte',
    'grid' : 'O2560', # currently O, F, N grids are supported 
    'type': 'fc',
}
```

```python
data_file = "data/climate-dt-earthkit-serverside-interpolation.grib"
if LIVE_REQUEST:
    data = earthkit.data.from_source("polytope", "destination-earth", request, address="polytope.lumi.apps.dte.destination-earth.eu", stream=False)
    data.to_target("file", data_file)
else:
    data = earthkit.data.from_source("file", data_file) 
```

```python
data.ls()
```

By default the interpolation method is linear and this can be specified with the `interpolation` key. The options available are `linera`, `bilinear`, `nearest neighbour`, `average`, and `nearest lsm`.

```python
request2 = {
    'activity': 'scenariomip',
    'class': 'd1',
    'dataset': 'climate-dt',
    'date': '20300101',
    'time': "0000",
    'experiment': 'SSP3-7.0',
    'expver': '0001',
    'generation': '1',
    'levtype': 'sfc',
    'model': 'IFS-NEMO',
    'param': '167',
    'realization': '1',
    'resolution': 'high',
    'stream': 'clte',
    'grid' : 'O1280', # currently O, F, N grids are supported 
    'interpolation': 'average',
    'type': 'fc',
}
```

```python
data_file2 = "data/climate-dt-earthkit-serverside-interpolation2.grib"
if LIVE_REQUEST:
    data2 = earthkit.data.from_source("polytope", "destination-earth", request2, address="polytope.lumi.apps.dte.destination-earth.eu", stream=False)
    data2.to_target("file", data_file2)
else:
    data2 = earthkit.data.from_source("file", data_file2) 
```
